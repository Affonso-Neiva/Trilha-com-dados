# Desepesas PMVV-2023

```python
dados_por_secretaria = pb.read_excel(r'C:\Users\patri\Desktop\Prefeitura_PANDAS\Despesa.Secretaria.xlsx')
```

```python
dados_por_secretaria
```

```python
dados_por_secretaria = dados_por_secretaria.iloc[:, 1:]
```

# Criando um novo dataframe apenas com as colunas desejadas:

```python
novascolunas = ['Nome','Orçado (R$)', 'Liquidado (R$)', 'Pago (R$)']
```

```python
tabela_secretaria = dados_por_secretaria.filter(items=novascolunas)
```

```python
tabela_secretaria.head()
```

# Mudando o Dtype das colunas:

```python
tabela_secretaria.info()
```

```python
tabela_secretaria['Orçado (R$)'] = tabela_secretaria['Orçado (R$)'].astype('int64')
```

```python
tabela_secretaria['Liquidado (R$)'] = tabela_secretaria['Liquidado (R$)'].astype('int64')
```

```python
tabela_secretaria['Pago (R$)'] = tabela_secretaria['Pago (R$)'].astype('int64')
```

```python
tabela_secretaria
```

# Plotando um gráfico de Barras lateral:

```python
tabela_secretaria.groupby('Nome').mean().plot.barh(stacked=True, figsize=(11,15))
```

# Criando um novo dataframe com as linhas desejadas e plotando um gráfico de pizza para comparação:

```python
df_secretaria = pb.DataFrame(
    {'Liquidado (R$)': [12496171, 77032314, 41964976, 1703736, 2532293], 
    'Orçado (R$)': [80673980, 627348417, 297854485, 19329959, 42770302]},
    index=['PROCURADORIA GERAL', 'SECR. MUN. DE EDUCAÇÃO', 'FUNDO MUNICIPAL DE SAÚDE',
           'SECR. MUN. DESENV. URBANO E MOBILIDADE', 'SECR. MUNICIPAL DE ADMINISTRAÇÃO']
)
```

```python
df_secretaria
```

```python
df_secretaria.plot.pie(y='Orçado (R$)', figsize=(10, 10))
```

```python
df_secretaria.plot.pie(y='Liquidado (R$)', figsize=(10, 10))
```

Em ambos os gráficos podemos ver que apesar da EDUCAÇÃO e SAÚDE terem as maiores despesas no ano de 2023, a PROCURADORIA GERAL ainda fica muito acima dos gastos do governo com DESENVOLVIMENTO URBANO E MOBILIDADE que geraria mais qualidade de vida a população canela verde.
