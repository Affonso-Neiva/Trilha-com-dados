# Construção do modelo OLS com Statsmodels em Python

```python
dados.head()
```

```python
# Definindo uma variável dependente:
y = dados['valor_aluguel']
```

```python
# definindo uma variável independente:
x = dados['area_m2']
```

```python
# O statsmodel requer que você adicione uma constante à variável independente:
x = sm.add_constant(x)
```

```python
# Criando o modelo:
modelo = sm.OLS(y, x)
```

```python
resultado = modelo.fit()
```

```python
print(resultado.summary())
```

# Interpretando 

```python
# Plot
plt.figure(figsize = (12, 8))
plt.xlabel('area_m2', size = 16)
plt.ylabel('valor_aluguel', size = 16)
plt.plot(x['area_m2'], y, "o", label = 'Dados Reais')
plt.plot(x['area_m2'], resultado.fittedvalues, "r-", label = 'Linha de Regressão (Previsões do modelo)')
plt.legend(loc = 'best')
plt.show()
```

# Conclusão

Claramente existe uma forte relação entre a area (m²) com o valor do aluguel. Entretando, apenas o valor da área não é suficiente para aplicar a variação no valor do aluguel, pois nosso modelo obteve um coeficiente de determinação (R²) de apenas 0,34 (quando o o máximo é 1 e quanto mais perto a ele melhor).

O ideal era utilizar mais de uma variável de entrada para construir o modelo a fim de compreender se outros fatores influenciam no valor do aluguel.

É sempre importande deixar claro que correlação não implica causalidade e que não podemos afirmar que o valor do aluguel mude devido a área dos imóveis
