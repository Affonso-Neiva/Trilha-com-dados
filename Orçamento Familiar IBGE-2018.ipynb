# % das famílias por região:

Separando o DF por Região e mudando o índice e deixando apenas dados numéricos:

```python
dados_por_regiao
```

```python
dados_por_regiao2 = dados_por_regiao.rename(
    index={5: 'Norte', 6: 'Nordeste', 7: 'Sudeste', 8:'Sul', 9: 'Centro-Oeste'}
)
```

```python
dados_por_regiao2 = dados_por_regiao2.iloc[:, 1:]
```

```python
dados_por_regiao2
```

Criando um novo DF com os mesmos índices e colunas:

```python
porcentagem_por_regiao = pd.DataFrame(
    columns=['Total','Até 1.908 Reais', 'Mais de 1.908 a 2.862 Reais', 'Mais de 2.862 a 5.724 Reais', 'Mais de 5.724 a 9.540 Reais', 'Mais de 9.540 a 14.310 Reais', 'Mais de 14.310 a 23.850 Reais', 'Mais de 23.850 Reais'],
    index=['Norte', 'Nordeste', 'Sudeste', 'Sul', 'Centro-Oeste']
)
```

```python
porcentagem_por_regiao
        
```

Preenchendo o DF usado FOR:

```python
for x in range(5):
    total = dados_por_regiao2.iloc[x, 0]
    for y in range(8):
        porcentagem_por_regiao.iloc[x, y] = (dados_por_regiao2.iloc[x, y] / total) * 100
```

```python
porcentagem_por_regiao = porcentagem_por_regiao.iloc[:, 1:]
porcentagem_por_regiao
```

```python
porcentagem_por_regiao.plot.barh();
```

# Dentro das análises, qual região se destaca nos extremos analisados?

```python
import matplotlib.pyplot as plt

figura = plt.figure()

#figura 1
figura.add_subplot(131)
ate_1908 = porcentagem_por_regiao['Até 1.908 Reais']
regioes = dados_por_regiao['Brasil, Grande Região e Unidade da Federação']
plt.title('Até 1.908 Reais')
plt.pie(ate_1908, labels = regioes)

#figura 2
figura.add_subplot(133)
Mais_de_23850 = porcentagem_por_regiao['Mais de 23.850 Reais']
plt.title('Mais de 23.850 Reais')
plt.pie(Mais_de_23850, labels = regioes)

plt.show()
```

Quando analisamos as faixas pelos extremos é nítida da dispariedade da concentração de renda por famílias. Na menor faixa analisada vemos o NORTE/NORDESTE se destacando com maior porcentagem das famílias. Na faixa maior faixa analizada vemos o CENTRO-OESTE se destacando dos demais.
