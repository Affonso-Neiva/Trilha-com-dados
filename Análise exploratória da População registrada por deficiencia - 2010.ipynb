# Deficiência visiual total por sexo:

```python
df_total = pd.DataFrame(dados_ibge, columns= ['Unnamed: 1', 'Unnamed: 2','Unnamed: 3','Unnamed: 4'])
```

```python
total = df_total['Unnamed: 2']
total = total[5]
total_m = df_total['Unnamed: 4']
total_m = total_m[5]
total_h = df_total['Unnamed: 3']
total_h = total_h[5]
```

# Renomeando as Colunas:

```python
df_total_grafico = df_total_grafico.rename(columns={'Unnamed: 1': 'Raça', 'Unnamed: 2': 'Total', 'Unnamed: 3': 'Homens', 'Unnamed: 4': 'Mulheres'}) 
```

```python
df_total_grafico
```

# Gerando um gráfico agrupado pela raça:

```python
df_total_grafico.groupby('Raça').mean().plot.barh(stacked=True);
```
